```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def q1():
   u = np.random.uniform(0, 1)
   inverse = np.power(1/(1-u), 1/3)
   
   return inverse


N = 100
result = np.zeros(N)

for i in range(N):
   result[i] = q1()


sample_average = sum(result)/len(result)
print("average == " + sample_average.astype(str))

plt.hist(result)
plt.title("Histogram of q1")
plt.show()
```

```python
def q2_itm():
   u = np.random.uniform(0, 1)
   inverse = np.log(1+(u*(np.e - 1)))
   return inverse


N = 1000
result = np.zeros(N)

for i in range(N):
   result[i] = q2_itm()


sample_average = sum(result)/len(result)
print("average == " + sample_average.astype(str))

plt.hist(result)
plt.title("Histogram of q2 via ITM")
plt.show()
```

```python
def q2_ar():

   def f(x):
      return np.exp(x)/(np.e-1)
   
   def g(x):
      return 1
   
   accept = False
   C = np.e/(np.e - 1)

   while (not accept):
      y = np.random.uniform(0, 1)
      u = np.random.uniform(0, 1)

      threshold = f(y)/(C*g(y))
      if (u <= threshold):
         return y
   
   return None


N = 1000
result = np.zeros(N)

for i in range(N):
   result[i] = q2_ar()


sample_average = sum(result)/len(result)
print("average == " + sample_average.astype(str))

plt.hist(result)
plt.title("Histogram of q2 via A-R")
plt.show()
```
