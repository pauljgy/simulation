```python
# problem 1 data
transition = np.array([[1/2, 1/3, 1/6],
              [0, 1/3, 2/3],
              [1/2, 0, 1/2]])



# 1c, 1d
M = 100

c = np.zeros(M)
d = np.zeros(M)
e = np.zeros(M)
f = np.zeros(M)

states = np.array([0, 1, 2])

for j in range(M):

    N = 100
    x = np.zeros(N)

    # initialization
    x[0] = 0 # np.random.choice(states)
    new = True

    # run the chain
    for i in range(1, N):
        state = int(x[i-1])
        x[i] = np.random.choice(states, p=transition[state])

        # f
        if (new and x[i] == 1):
            f[j] = i
            new = False
    
    #1c
    if (x[10] == 1):
        c[j] = 1

    #1d
    d[j] = x[10]

    #1e
    if (1 in x[0:11]):
        e[j] = 1


#1c
print(f"1c point estimate: {np.sum(c)/M}")

#1d
print(f"1d point estimate: {np.sum(d)/M}")

#1e
print(f"1e point estimate: {np.sum(e)/M}")

#1f
print(f"1f point estimate: {np.sum(f)/M}")
```

```python
N = 10000
x = np.zeros(N)

# initialization
x[0] = 0
states = np.array([0, 1, 2])

# run the chain
for i in range(1, N):
    state = int(x[i-1])
    x[i] = np.random.choice(states, p=transition[state])


visits = np.zeros(3)
costs = np.zeros(N-100)
cost = {0: 1,
        1: 3,
        2: 2}

for i in range(100, N):
    state = int(x[i])
    visits[state] += 1
    costs[i-100] = cost[state]

print(f"1e visit proportion: {np.divide(visits, (N-100))}")
print(f"1f average cost: {np.sum(costs)/(N-100)}")
```

```python
# problem 2

# 2b
states = np.array([1, 0])
trans = np.array([[0.2, 0.8],
                  [0.6, 0.4]])
N = 100
T = 10000

x = np.zeros(T)
x[0] = 10

for i in range(1, T):
    state = int(x[i-1])

    active = 0

    for j in range(N):
        if (j < state):
            active += np.random.choice(states, p=trans[0])
        else:
            active += np.random.choice(states, p=trans[1])
    
    x[i] = active
    
print(f"2b Avg number of actives: {np.sum(x[1000:T])/9000}")
```

```python
# 2d
states = np.array([1, 0])
trans = np.array([[0.2, 0.8],
                  [0.6, 0.4]])

N = 100
T = 10000

# y = active and initially active.
# z = active and intiially not active.
y = np.zeros(T)
z = np.zeros(T)

y[0] = 10
z[0] = 0

for i in range(1, T):
    y_state = int(y[i-1])
    z_state = int(z[i-1])

    y_active = 0
    for j in range(0, 10):
        if (j < y_state):
            y_active += np.random.choice(states, p=trans[0])
        else:
            y_active += np.random.choice(states, p=trans[1])
    y[i] = y_active

    z_active = 0
    for j in range(0, 90):
        if (j < z_state):
            z_active += np.random.choice(states, p=trans[0])
        else:
            z_active += np.random.choice(states, p=trans[1])
    z[i] = z_active

y_wins = np.zeros(9000)
for i in range(9000):
    y_wins = (1 if y[i+1000] > z[i+1000] else 0)
print(f"2b probability of y_n > z_n: {np.sum(y_wins)/9000}")
```
