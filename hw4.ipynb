```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def q2():
    mean = [0, 0]
    cov = [[1, 0], [0, 1]]

    accept = False
    
    while (not accept):
        g = np.random.multivariate_normal(mean, cov)
        u = np.random.uniform(0, 1)

        x = g[0]
        y = g[1]
        h = np.exp(-x**2 - y**2 + (x * np.sin(x*y)))
        Cg = np.exp(-1/2 * (x**2 + y**2 - 1))

        if (u < h/Cg):
            return g
    
    return None


N = 10000
x_data = np.zeros(N)
y_data = np.zeros(N)

for i in range(N):
    tx = q2()
    x_data[i] = tx[0]
    y_data[i] = tx[1]
```

```python
plt.scatter(x_data, y_data, s=1)
plt.scatter(0, 0, s=5, c="red")
plt.title("simulating joint pdf with unknown normalization constant")
plt.show()
```

```python
def itm():
    u = np.random.uniform(0, 1)**2

    return u

N = 10000
u_test = np.zeros(N)

for i in range(N):
    u_test[i] = itm()
    
plt.hist(u_test)
plt.show()
```

```python
def q3():
    sign = np.random.uniform(0, 1)
    sign = 1 if (sign > 0.5) else -1

    u = itm() * sign

    y = np.random.uniform(-(np.abs(u)**(-1/2)), np.abs(u)**(-1/2))
    
    return u, y

N = 10000
x_data = np.zeros(N)
y_data = np.zeros(N)

for i in range(N):
    temp = q3()
    x_data[i] = temp[0]
    y_data[i] = temp[1]
```

```python
plt.scatter(x_data, y_data, s=1)
plt.ylim(-10, 10)
plt.title("uniform sampling from irregular region")
plt.show()
```

```python
def q4():
    choice = 1 if (np.random.uniform(0,1) > 0.5) else 0
    x = 0
    if (choice):
        x = np.random.geometric(1/2)
    else:
        x = np.random.geometric(1/3)

    return x

N = 1000
test = np.zeros(N)

for i in range(N):
    test[i] = q4()

plt.hist(test, bins=15)
plt.title("HW2 Problem 3 via composition method")
plt.show()
```

```python
def q5():
    y = np.random.exponential(1)
    u = np.random.uniform(0, 1)
    x = np.power(u, 1/y)
    return x

N = 100000
u_test = np.zeros(N)

for i in range(N):
    u_test[i] = q5()
    
plt.hist(u_test, bins=20)
plt.show()
```
